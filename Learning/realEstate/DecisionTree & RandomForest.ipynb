```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

train=pd.read_csv("../data/realEstate/train.csv")
test=pd.read_csv("../data/realEstate/test.csv")
train.fillna(0,inplace=True)
test.fillna(0,inplace=True)
train.describe()
```

```python
train.isnull().sum().sum()
```

```python
y=train.SalePrice
X=train.select_dtypes(include=[np.number])
X.head()
X.remove('Id','SalePrice')
```

```python
X=X.drop(['Id', 'SalePrice'], axis=1)
```

```python
realEstate_model = DecisionTreeRegressor(random_state=1)
realEstate_model.fit(X,y)
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(realEstate_model.predict(X.head()))
```

```python
predicted_home_prices = realEstate_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
training_X,validation_X,training_y,validation_y=train_test_split(X,y,random_state=0)

def get_mae(max_leafs, training_X,validation_X,training_y,validation_y):
    model_reg = DecisionTreeRegressor(max_leaf_nodes=max_leafs, random_state=0)
    model_reg.fit(training_X,training_y)
    predicted_model_reg = model_reg.predict(validation_X)
    mae=mean_absolute_error(validation_y, predicted_model_reg)
    return mae
for max_leafs in [5,50,100,500,5000,10000]:
    mae=get_mae(max_leafs, training_X,validation_X,training_y,validation_y)
    print(max_leafs, '-----',mae)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(training_X,training_y)
pred_forest_model = forest_model.predict(validation_X)
print(mean_absolute_error(validation_y, pred_forest_model))
```
